### Approximation non-lineaire

```python
pany2pan("lena.bmp","lena.pan")
pqmf("daubechies",4,"qmf.pan")
pdwt(5,"lena.pan","qmf.pan","lenaTO.pan")
pvisuJG("lenaTO.pan")
```

## DWT

```python
pim2sf("lena.pan","lena.pan")
pabs("lenaTO.pan","lenaABS.pan")

energy = []
X = [k for k in range(0,100)]

for x in X:
     
    pmassbinarization(x,"lenaABS.pan","lenamask.pan")
    
    pmask("lenaTO.pan","lenamask.pan","lenamasked.pan")
    
    pidwt(5,"lenamasked.pan","qmf.pan","lenamaskedInv.pan")
    
    psub("lena.pan","lenamaskedInv.pan","lenaDiff.pan")
    
    penergyvalue("lenaDiff.pan","lenaEnergy.pan")
    
    energy.append(float(pstatusJ()))
```

```python

rev = X[::-1]

plt.plot(rev,energy)

plt.yscale("linear")
```

## FFT

```python
pany2pan("lena.bmp","lena.pan")
pim2sf("lena.pan","lena.pan")
psetcst(0,"lena.pan","lena_im.pan")

pfft("lena.pan","lena_im.pan","lena_fft_real.pan","lena_fft_im.pan")
pmodulus("lena_fft_real.pan","lena_fft_im.pan","lena_modulus.pan")
```

```python
energy2 = []
X = [k for k in range(0,100)]

for x in X:
     
    pmassbinarization(x,"lena_modulus.pan","lenamask.pan")
    
    pmask("lena_fft_real.pan","lenamask.pan","lenamasked.pan")
    pmask("lena_fft_im.pan","lenamask.pan","lenamasked_im.pan")
    
    pifft("lenamasked.pan","lenamasked_im.pan","lena_ifft.pan","lena_ifft_im.pan")
    
    psub("lena.pan","lena_ifft.pan","lenaDiff.pan")
        
    penergyvalue("lenaDiff.pan")
    
    energy2.append(float(pstatusJ()))
```

```python

rev = X[::-1]

plt.plot(rev,energy2)
plt.plot(rev,energy)

plt.yscale("linear")
```

### DEBRUITAGE

```python
pany2pan("lena.bmp","lena.pan")

pim2sf("lena.pan","lena.pan")

paddnoise(1,0,20,"lena.pan","lena_noise.pan")

pqmf("daubechies",4,"qmf.pan")

pdwt(3,"lena_noise.pan","qmf.pan","lena_noise_TO.pan")

pgetsubband(3,1,"lena_noise_TO.pan","lenasubband.pan")

pabs("lena_noise_TO.pan","lena_noise_TO_ABS.pan")

pvisuJG("lena_noise_TO.pan")
```

```python
lbda = 20*mp.sqrt(2*mp.log(256*256))

pbinarization(lbda,255,"lena_noise_TO_ABS.pan","lenamask.pan")

pmask("lena_noise_TO.pan","lenamask.pan","lenamasked.pan")

psetsubband(3,1,"lenamasked.pan","lenasubband.pan","lenambanded.pan")

pidwt(3,"lenambanded.pan","qmf.pan","lenaInv.pan")

pvisuJG("lenaInv.pan")
```

```python
lbd = 3*20

pbinarization(lbda,255,"lena_noise_TO_ABS.pan","lenamask.pan")

pmask("lena_noise_TO.pan","lenamask.pan","lenamasked.pan")

psetsubband(3,1,"lenamasked.pan","lenasubband.pan","lenambanded.pan")

pidwt(3,"lenambanded.pan","qmf.pan","lenaInv2.pan")

pvisuJG("lenaInv2.pan")
```

## influence du seuil

```python
tresholding = [0,30,70,100,150,200]

for v in tresholding :
    print(v)
    pbinarization(v,255,"lena_noise_TO_ABS.pan","lenamask.pan")

    pmask("lena_noise_TO.pan","lenamask.pan","lenamasked.pan")

    psetsubband(3,1,"lenamasked.pan","lenasubband.pan","lenambanded.pan")

    pidwt(3,"lenambanded.pan","qmf.pan","lenaInv2.pan")

    pvisuJG("lenaInv2.pan")
```

## influences de la variance du bruit et de lâ€™echelle de decomposition de la TO

```python
energy = []
X = [k for k in range(0,250)]

for x in X:
    pbinarization(x,255,"lena_noise_TO_ABS.pan","lenamask.pan")

    pmask("lena_noise_TO.pan","lenamask.pan","lenamasked.pan")

    psetsubband(3,1,"lenamasked.pan","lenasubband.pan","lenambanded.pan")

    pidwt(3,"lenambanded.pan","qmf.pan","lenaInv.pan")
    
    psub("lena.pan","lenaInv.pan","lenaDiff.pan")
    
    penergyvalue("lenaDiff.pan")
    
    energy.append(float(pstatusJ()))
    
```

```python
plt.plot(X,energy)

optimal = X[energy.index(min(energy))]
print("l'optimal est : ", optimal)
```

## Influence du noise

```python
S = [10,20,40,70]

pany2pan("lena.bmp","lena.pan")

pim2sf("lena.pan","lena.pan")

pqmf("daubechies",4,"qmf.pan")


for s in S :
    paddnoise(1,0,s,"lena.pan","lena_noise.pan")

    pdwt(3,"lena_noise.pan","qmf.pan","lena_noise_TO.pan")

    pgetsubband(3,1,"lena_noise_TO.pan","lenasubband.pan")

    pabs("lena_noise_TO.pan","lena_noise_TO_ABS.pan")
    
    energy = []
    X = [k for k in range(0,250)]

    for x in X:
        pbinarization(x,255,"lena_noise_TO_ABS.pan","lenamask.pan")

        pmask("lena_noise_TO.pan","lenamask.pan","lenamasked.pan")

        psetsubband(3,1,"lenamasked.pan","lenasubband.pan","lenambanded.pan")

        pidwt(3,"lenambanded.pan","qmf.pan","lenaInv.pan")

        psub("lena.pan","lenaInv.pan","lenaDiff.pan")

        penergyvalue("lenaDiff.pan")

        energy.append(float(pstatusJ()))
    
    plt.plot(X,energy)

    optimal = X[energy.index(min(energy))]
    print("l'optimal est : ", optimal)
```

```python
S = [1 ,10 ,20,40,60]

pany2pan("lena.bmp","lena.pan")

pim2sf("lena.pan","lena.pan")

pqmf("daubechies",4,"qmf.pan")

energy = []

for s in S :
    paddnoise(1,0,s,"lena.pan","lena_noise.pan")

    pdwt(3,"lena_noise.pan","qmf.pan","lena_noise_TO.pan")

    pgetsubband(3,1,"lena_noise_TO.pan","lenasubband.pan")

    pabs("lena_noise_TO.pan","lena_noise_TO_ABS.pan")
    
    pbinarization(3*s,255,"lena_noise_TO_ABS.pan","lenamask.pan")
    
    pmask("lena_noise_TO.pan","lenamask.pan","lenamasked.pan")

    psetsubband(3,1,"lenamasked.pan","lenasubband.pan","lenambanded.pan")

    pidwt(3,"lenambanded.pan","qmf.pan","lenaInv.pan")
    
    psub("lena.pan","lenaInv.pan","lenaDiff.pan")
    
    penergyvalue("lenaDiff.pan")
    
    energy.append(float(pstatusJ()))
    
plt.plot(S,energy)
```
